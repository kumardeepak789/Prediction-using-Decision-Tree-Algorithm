## importing Libraries

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

## Reading the dataset

```python
df=pd.read_csv("iris.csv",index_col=0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

## Extracting the features from dataset

```python
X=df.iloc[:,0:4]
```

## Dividing The Dataset into Training(75%) and Testing(25%) data

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,0:4],df.iloc[:,4],train_size=0.75,random_state=2)
```

```python
decision_tree=DecisionTreeClassifier()
```

## Training the model

```python
decision_tree.fit(xtrain,ytrain)
```

## The accuracy of the model

```python
print("The accuracy of model is ",decision_tree.score(xtest,ytest))
```

## visuallizing tree

```python
names=[]
names=df.columns
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(25,20))
plot_tree(decision_tree,feature_names=names[[0,1,2,3]],class_names=names[4],filled=True)
```
